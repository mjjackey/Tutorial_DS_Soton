**R**
This is a small exercise in using R. The aim is to analyse a data set of catches from a fishing fleet. 

The dataset "fish1.txt" records data for a time period of one day during  which one fisherman has fished in a lake. The fisherman uses three types of fishing  rods, labeled A, B, and C, each using different bait. The fisherman has recorded every  catch he has made during this time. The data set consists of three columns with X  values giving the times at which the fisherman has made a catch, the Y values indicate the size of that catch (i.e. its weight in kg), and the Z values give a letter A, B, or C which indicates which fishing rod was used to make that catch.

The first task is to import the dataset:

```r
fishdata <- read.table("fish1.txt")
```

where:  
fishdata is the placeholder for the data  
<- is the assignment operator

**Plotting in R**

To find the variable names use: names(fishdata)  
Then attach themto be available for use in R commands: attach(fishdata)  
To plot variable1 against variable2 use: plot(variable1 ~ variable2)  
To export as a pdf: pdf("filename.pdf")
