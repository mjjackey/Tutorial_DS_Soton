```r
randNum<-c(10,100,1000)
meanArray<-double(3)
j<-0
for (i in randNum)  #10100100010000
{mySample <-sample(rnorm(10000), i, replace=F) #sample,isamplesize,replace=T 
 j<-j+1
 print (c(i, meanArray[j]<-mean(mySample),sd(mySample))) } 
```
